```python
from bs4 import BeautifulSoup
import requests
```

```python
url_games='https://soccer365.ru/competitions/12/results/'
get_request = requests.get(url_games, timeout=10, params=None, headers=None, data=None)
```

    Создание объекта BeautifulSoup
    
    Назначение: Парсинг HTML/XML для дальнейшей работы с деревом элементов.
    
    Синтаксис:
                from bs4 import BeautifulSoup
                soup = BeautifulSoup(html_content, 'html.parser')  # или 'lxml', 'html5lib'
    
    Параметры:
                html_content: Строка или файл с HTML/XML.
                Парсер (html.parser, lxml, html5lib): Выбор зависит от скорости и точности.
    
    Пример:
            html = "<html><body><p class='text'>Hello World</p></body></html>"
            soup = BeautifulSoup(html, 'html.parser')

    Когда использовать: Всегда в начале работы для создания объекта-парсера.

```python
soup = BeautifulSoup(markup=get_request.text, features="html.parser")   # или 'lxml', 'html5lib'
print(soup)
```

    find()
    Назначение: Поиск первого элемента, соответствующего условию.

    Синтаксис:
            element = soup.find(name, attrs={}, string=None)

    Параметры:
            name: Название тега (например, 'div', 'a').
            attrs: Словарь атрибутов (например, {'class': 'text'}).
            string: Текст внутри тега.

    Пример:
            p_tag = soup.find('p', attrs={'class': 'text'})  # Находит <p class="text">...</p>

    Когда использовать: Когда нужен первый элемент с определёнными атрибутами или текстом.

```python
find_first_link_tag = soup.find(name="link", attrs={"hreflang": "en"}, string=None)
print(find_first_link_tag)
```

    find_all()
    
    Назначение: Поиск всех элементов, соответствующих условию.

    Синтаксис:
            elements = soup.find_all(name, attrs={}, string=None, limit=None)
    
    Параметры:
            limit: Максимальное количество элементов для возврата.

    Пример:
            all_links = soup.find_all('a', attrs={'target': '_blank'})  # Все ссылки с target="_blank"
    
    Когда использовать: Для сбора списка элементов (например, всех ссылок на странице).

```python
find_all_a_tag = soup.find_all(name="a", string=None, limit=5)
print(*find_all_a_tag, sep="\n")
```

    select() (CSS-селекторы)
    
    Назначение: Поиск элементов по CSS-селектору.

    Синтаксис:
            elements = soup.select('css_selector')
    
    Примеры селекторов:
            'div.content' → элементы <div class="content">.
            'a#link' → элемент <a id="link">.
            'ul > li' → все <li>, являющиеся прямыми потомками <ul>.

    Пример:
            items = soup.select('ul.menu > li')  # Все <li> внутри <ul class="menu">

    Когда использовать: Для сложных запросов, где удобнее использовать CSS-синтаксис.

```python
items = soup.select("a.icon16", limit=0)
print(*items, sep="\n")
```

    get_text()
    
    Назначение: Извлечение текста из элемента (включая вложенные теги).

    Синтаксис:
            text = element.get_text(separator=' ', strip=False)
    
    Параметры:
            separator: Разделитель для текста разных тегов.
            strip: Удаление пробелов в начале/конце.
    
    Пример:
            p_text = p_tag.get_text(strip=True)  # "Hello World"

    Когда использовать: Для извлечения «чистого» текста без HTML-тегов.

```python
a_text = soup.find(name="a", string="Общение")
for a in find_all_a_tag:
    text = a.get_text(strip=True, separator="")
    print(text if text else "Текст данного тега отсутствует")
text = a_text.get_text(separator=" ", strip=True)
print(text)
```

    Работа с атрибутами тегов
    
    Назначение: Доступ к атрибутам тега (например, href, class, id).

    Синтаксис:
            value = tag['attribute']          # Получить атрибут
            tag['attribute'] = 'new_value'    # Изменить атрибут
    
    Пример:
            link = soup.find('a')
            url = link['href']  # Получить значение href
    
    Когда использовать: Для извлечения ссылок, классов, идентификаторов и других атрибутов.

```python
links = soup.find_all(name="link", limit=5)
for link in links:
    url = link["href"]
    print(f"{url = }")
link["href"] = "No link"
print(f"{link["href"] = }")
link["href"] = "new_link"
print(f"{link["href"] = }")
```

    parent, children, next_sibling, previous_sibling, descendants
    
    Назначение: Навигация по дереву элементов.

    Методы:
            parent: Родительский элемент.
            children: Прямые потомки (генерирует элементы по одному).
            next_sibling / previous_sibling: Соседние элементы того же уровня.
            descendants — все вложенные элементы рекурсивно.
                Есть методы `find_next_sibling()` и `find_previous_sibling()`, которые позволяют искать соседей, 
                удовлетворяющих определенным условиям, в отличие от простых `next_sibling`, 
                которые просто возвращают следующий элемент независимо от его типа.

    Пример:
            parent_div = link.parent          # Родительский тег ссылки
            first_child = parent_div.children[0]  # Первый потомок
    
    Когда использовать: Для перемещения по структуре HTML (например, переход к родителю или соседнему элементу).

```python
parent_div = link.parent
first_child = parent_div.children
print(next(first_child))


```

```python
html = '''
<html>
  <body>
    <div class="container">
      <p>Пример текста</p>
    </div>
  </body>
</html>
'''
```

```python
# parent — Родительский элемент
# Что делает: Возвращает непосредственного родителя текущего тега.
# Особенности:
#   Если элемент находится в корне (например, тег <html>), его parent — это объект BeautifulSoup.
#   У строк текста (не тегов) тоже есть родители.
soup = BeautifulSoup(markup=html, features="html.parser")
p_tag = soup.find(name="p")  # Находим тег <p>

# Получаем родителя <p> — это <div class="container">
parent = p_tag.parent
print(f"{parent.name = }")
print(parent.prettify())
# Когда использовать:
#   Чтобы найти контейнер, в котором находится элемент.
#   Для подъема на уровень выше в структуре HTML.
```

```python
# children — Прямые потомки
# Что делает: Возвращает генератор с прямыми потомками тега (только первого уровня вложенности).
# Особенности:
#   Включает не только теги, но и строки текста, пробелы и переносы строк.
#   Чтобы получить список элементов, нужно преобразовать генератор в список: list(parent.children).
soup = BeautifulSoup(markup=html, features="html.parser")
div_tag = soup.find(name="div")

# Перебираем всех прямых потомков div
for child in div_tag.children:
    print(repr(child))
print()
    
# Как отфильтровать только теги:
real_children = [child for child in div_tag.children if child.name is not None] # child.name -- дает имя тега
print(real_children)

# Когда использовать:
#   Для анализа непосредственных дочерних элементов.
#   Когда нужно обработать только первый уровень вложенности.
```

```python
# next_sibling и previous_sibling — Соседние элементы

# Что делает:
#   next_sibling — возвращает следующий элемент на том же уровне вложенности.
#   previous_sibling — возвращает предыдущий элемент.

# Особенности:
#   Часто возвращают пробелы или переносы строк (так как они тоже считаются элементами).

#   Чтобы пропустить "мусорные" элементы, используйте find_next_sibling() или find_previous_sibling().
html = '''
<ul>
  <li>Пункт 1</li>
  <li>Пункт 2</li>
  <li>Пункт 3</li>
</ul>
'''
soup = BeautifulSoup(markup=html, features="html.parser")
first_li = soup.find(name="li")

# Следующий сосед (учтите пробелы и переносы строк!)
next_sibling_li = first_li.next_sibling
print(repr(next_sibling_li))

# Чтобы получить следующий тег <li>, используйте find_next_sibling()
real_next = first_li.find_next_sibling(name="li")
print(real_next)
real_next_tags = first_li.find_next_siblings(name="li")
print(real_next_tags)

# Когда использовать:
#   Для парсинга списков, таблиц или других структур, где элементы идут последовательно.
#   Когда нужно переходить к соседним элементам на одном уровне.
```

```python
# find_next_sibling() и find_previous_sibling()
# Что делают: Ищут следующего/предыдущего соседа, удовлетворяющего условиям.
# Синтаксис:
#   next_li = tag.find_next_sibling(name='li', attrs={'class': 'active'})
html = '''
<div>
  <p class="info">Текст 1</p>
  <p class="warning">Текст 2</p>
  <p class="info">Текст 3</p>
</div>
'''
soup = BeautifulSoup(markup=html, features="html.parser")
first_p = soup.find(name="p", class_="info")
print(first_p)

# Ищем следующий тег <p> с классом 'info'
new_info_p = first_p.find_next_sibling(name="p",  class_="info")
print(new_info_p.text)
```

```python
# parents — Все родители
# Что делает: Возвращает генератор всех родителей элемента (от непосредственного до корня).
# Пример:
soup = BeautifulSoup(markup=html, features="html.parser")
p_tag = soup.find(name="p")
# Выводим всех родителей тега <p>
for parent in p_tag.parents:
    print(parent.name)
# Вывод: div -> body -> html -> [document]
```

```python
# descendants — Все потомки (рекурсивно)
# Отличие от children: descendants включает все вложенные элементы (не только первого уровня).
# Пример:
soup = BeautifulSoup(markup=html, features="html.parser")
div_tag = soup.find(name="div")

# Перебираем всех потомков div (теги и текст)
for child in div_tag.descendants:
    print(repr(child))
```

```python
# Практические советы
#   1.Обработка пробелов и переносов:
# Пропускаем строки-невидимки
soup = BeautifulSoup(markup=html, features="html.parser")
tag = soup.find(name="p")
real_siblings = [sib for sib in tag.next_siblings if sib.name is not None]
print(real_siblings)

#   2.Проверка на None:
if tag.parent is not None:
    print(tag.parent.name)
    
#   3.Цепочки вызовов:
# Переход от ребенка к родителю и поиск другого тега
title = tag.parent.find('h1')
print(title)
```

```python
# Пример: Парсинг таблицы
with open(file="index.html", mode="rt", encoding="utf-8") as file:
    html = file.read()
    
soup = BeautifulSoup(markup=html, features="html.parser")
# Находим первую ячейку
first_td = soup.find(name="td")

# Переходим к родителю <tr>, затем ищем все <td> внутри него
first_row = first_td.parent
cells = first_row.find_all(name="td")
first_cells = [cell.text for cell in cells]

second_row = first_row.find_next_sibling()
cells = second_row.find_all(name="td")
second_cells = [cell.text for cell in cells]
print(*first_cells, sep=" | ")
print("-" * 20)
print(*second_cells, sep=" | ")
```

    has_attr()
    
    Назначение: Проверка наличия атрибута у тега.

    Синтаксис:
            if tag.has_attr('class'):
                print("Тег имеет класс")
    
    Когда использовать: Для безопасной проверки атрибутов, чтобы избежать ошибок KeyError

```python
soup = BeautifulSoup(markup=get_request.text[100:], features="html.parser")
a_tag = soup.find(name="a", string=None)
print(f"{a_tag.has_attr("href") = }")
print(f"{a_tag.has_attr("url") = }")
```

    decompose() и extract()
    
    Назначение: Удаление элементов из дерева.

    Различия:
            decompose(): Удаляет элемент и полностью уничтожает его.
            extract(): Удаляет элемент, но сохраняет его для дальнейшего использования.

    Пример:
            unwanted_tag = soup.find('script')
            unwanted_tag.decompose()  # Удалить скрипт из документа
    
    Когда использовать: Для очистки HTML от ненужных тегов (например, скриптов).

```python
deleted_tag = soup.find("script")
deleted_tag.extract()
# P.S.:
#   Я все удалил)
```

    Пример:
            print(soup.prettify())  # Выводит аккуратно отформатированный HTML
    
    Когда использовать: Для отладки или сохранения «красивого» HTML.

```python
pretty_html = soup.prettify()
print(pretty_html)
```
